```python
import torch.nn as nn 
import torch.nn.functional as F
import numpy as np
```

```python
def loss(student_result, teacher_hard_result, teacher_soft_result, temp=1.2, alpha=0.7):
    # hard loss is cross entropy
    hard_loss = nn.CrossEntropyLoss()(student_result, teacher_hard_result) 
    
    # soft loss is kl divergence loss
    student_prob = F.log_softmax(student_result/temp, dim=1)
    teacher_prob = F.softmax(teacher_soft_result/temp, dim=1)
    soft_loss = nn.KLDivLoss(reduction='batchmean')(student_prob, teacher_prob) * (temp ** 2)

    # combine loss with alpha
    total_loss = alpha * soft_loss + (1 - alpha) * hard_loss 
    return total_loss
```
